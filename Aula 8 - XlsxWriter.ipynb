```python
import xlsxwriter as xlsx
import os

#------------------------------------------------------Caminho e aquivo
caminho = 'C:\\Users\\FIC\\Documents\\Emerson-Python\\'
arquivo = 'Desafio.xlsx'

#------------------------------------------------------Criar um arquivo e uma aba
workbook = xlsx.Workbook(caminho + arquivo)
sheetDados = workbook.add_worksheet('Dados')

#------------------------------------------------------Modificando a cor de fundo das células
corFonte = workbook.add_format()
corFonte.set_font_color('blue')

#Cabeçalho
corFundoCabeçalho = workbook.add_format({
                                'bold':True, #Negrito
                                'border':True #aplicar bordas
                               })

#Títulos
corFundoTitulos = workbook.add_format({
                                'align': 'center', #Alinhar textos
                                'fg_color':'#F4D0C4',  #cor do preenchimento 
                                'font_color':'#1D3857', #corda fonte
                                'bold': True, #Negrito
                                'border': True #aplicar bordas
                               })

#Totais
corFundoTotais = workbook.add_format({
                                'fg_color':'#F4D0C4',  #cor do preenchimento 
                               # 'font_color':'#1D3857', #corda fonte
                                'bold':True, #Negrito
                                'border':True #aplicar bordas
                              #  'num_formart':'R$ #,##0' #Alterar formato
                               })


#Sub-Titulos
corFundoSub = workbook.add_format({
                                
                                'fg_color':'#EFF6CC',  #cor do preenchimento
                                'bold':True, #Negrito
                                'border':True, #aplicar bordas
                                'align': 'center' #Alinhar textos
                               })

#Dados
corFundoDados = workbook.add_format({

                                'border':True, #aplicar bordas
                                'num_format':'R$ #,##0' #Alterar formato
                               })


#------------------------------------------------------Escrever nas células
sheetDados.merge_range('A1:C1','Controle de Custos',corFundoCabeçalho)
sheetDados.merge_range('A2:A3','Itens',corFundoTitulos)
sheetDados.write('A4','Salário',corFundoSub)
sheetDados.write('A5','Horas Extras',corFundoSub)
sheetDados.write('A6','Previdência Social',corFundoSub)
sheetDados.write('A7','FGTS',corFundoSub)
sheetDados.write('A8','Aluguel',corFundoSub)
sheetDados.write('A9','Luz/Água/Telefone',corFundoSub)
sheetDados.write('A10','Mat.Expediente',corFundoSub)
sheetDados.write('A11','Outras Despesas',corFundoSub)
sheetDados.write('A12','Total',corFundoTotais)

sheetDados.merge_range('B2:C2','Total 1º Trimestre',corFundoTitulos)
sheetDados.write('B3','Previsão',corFundoTitulos)
sheetDados.write('B4',2000,corFundoDados)
sheetDados.write('B5',3150,corFundoDados)
sheetDados.write('B6',4300,corFundoDados)
sheetDados.write('B7',5450,corFundoDados)
sheetDados.write('B8',6600,corFundoDados)
sheetDados.write('B9',7750,corFundoDados)
sheetDados.write('B10',8900,corFundoDados)
sheetDados.write('B11',10050,corFundoDados)
sheetDados.write('B12','=sum(B4:B11)',corFundoTotais)

sheetDados.write('C3','Gastos',corFundoTitulos)
sheetDados.write('C4',1550,corFundoDados)
sheetDados.write('C5',2550,corFundoDados)
sheetDados.write('C6',3550,corFundoDados)
sheetDados.write('C7',4550,corFundoDados)
sheetDados.write('C8',5550,corFundoDados)
sheetDados.write('C9',6550,corFundoDados)
sheetDados.write('C10',7550,corFundoDados)
sheetDados.write('C11',8550,corFundoDados)
sheetDados.write('C12','=sum(C4:C11)',corFundoTotais)

#------------------------------------------------------Fechar o documento e salvar
workbook.close()

#------------------------------------------------------Abrir o documento
os.startfile(caminho + arquivo)

```
