```python
#----------------------------------------Importação das Bibliotecas
import xlsxwriter as xlsx
import os

#----------------------------------------caminho e arquivo
caminho = 'C:\\Users\\FIC\\Documents\\Emerson-Python\\'
arquivo = 'Exercicio_5.xlsx'

#----------------------------------------criar pasta de trabalho
workbook = xlsx.Workbook(caminho + arquivo) #arquivo do excel
sheetDados = workbook.add_worksheet('Relatório') #Planilha

#----------------------------------------listar os dados
dados = [
    ['Produto','Qtde','Preço','Subtotal'],
    ['Calça',5,50],
    ['Camisa',4,30],
    ['Sapato',5,40],
    ['Saia',8,20],
    ['Blusa',6,10],
    ['Shorts',8,25]
]

#----------------------------------------Formatação------------------  
#Títulos
titles = workbook.add_format({
                                'bg_color':'blue',
                                'font_color':'white',
                                'font_name':'Arial',
                                'align':'center',
                                'bold':'True'    
})    

#rotulos
rotulosLinhas = workbook.add_format({
                                        'font_color':'gray',
                                        'font_name':'Arial',
                                        'align':'left'
})

#quantidade
qtde = workbook.add_format({
                                'font_color':'gray',
                                'font_name':'Arial',
                                'align':'center'
})

#Moedas
valores = workbook.add_format({
                                'font_color':'gray',
                                'font_name':'Arial',
                                'align':'center',
                                'num_format':'R$ * #.##0,0'
})
#formato de data
formatoData = workbook.add_format({
                                        'font_color':'gray',
                                        'font_name':'Arial',
                                        'align':'center',
                                        'num_format':'dd/mm/yyyy'    
})
#formato de hora
formatoHora = workbook.add_format({
                                        'font_color':'gray',
                                        'font_name':'Arial',
                                        'align':'center',
                                        'num_format':'h:mm'    
})

#mesclagem
mesclado = workbook.add_format({
                                    'font_color':'black',
                                    'font_name':'Arial',
                                    'font_size':22,
                                    'align':'center',
                                    'bold':True
})

 #Cabeçalho
corFundoCabeçalho = workbook.add_format({
                                'align': 'center', #Alinhar textos
                                'bold':True, #Negrito
                                'border':True #aplicar bordas
                               })

#----------------------------------------Registrar data
sheetDados.merge_range('A1:D1', 'Planilha de Solicitações',mesclado)

sheetDados.write('A3','Data')
sheetDados.write_formula('B3','=Today()',formatoData)
sheetDados.write('A4','Hora')
sheetDados.write_formula('B4','=Now()',formatoHora)

#----------------------------------------Registras os dados na planilha
posLinha = 5
posColuna = 0

for linha, listas in enumerate(dados):
    if linha ==0:
        sheetDados.write_row(linha + posLinha,posColuna,listas,titles) #cabeçalhos
    else:
        sheetDados.write(linha + posLinha,posColuna,listas[0]) #Produto
        sheetDados.write(linha + posLinha,posColuna,listas[1]) #Quantidade
        sheetDados.write(linha + posLinha,posColuna,listas[2],valores) #Preço unitário
        sheetDados.write_formula(linha + posLinha,posColuna + 3,'=B'+ str(linha + posLinha+1) + '*C' + str(linha + posLinha+1) )


#----------------------------------------Gerar Total
sheetDados.write('D13','Total')
sheetDados


sheetDados.merge_range('A3:D3','Planilha de Solicitações',corFundoCabeçalho)

#----------------------------------------fechar o arquivo e salvar

sheetDados.set_column('A:D',15.5)
workbook.close()

#----------------------------------------Abrir documento
os.startfile(caminho + arquivo)

```
